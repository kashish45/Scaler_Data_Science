# BAR PLOT

```python
sales_swift=[13,19,25,11,15]
sales_nano=[24,19,15,11,8] #We have created lists
```

```python
sales_nano=np.array(sales_nano)
sales_swift=np.array(sales_swift)
```

```python
x=np.arange(5)
```

```python
plt.bar(x,sales_swift)
plt.show()
```

```python
plt.bar(x,sales_swift,width=0.3)
plt.bar(x,sales_nano,width=0.3)
plt.show()
```

```python
plt.bar(x,sales_swift,width=0.3)
plt.bar(x,sales_nano,width=0.3,alpha=0.7)
plt.show()
```

```python
plt.bar(x,sales_swift,width=0.3,color='red')
plt.bar(x,sales_nano,width=0.3,alpha=0.7)
plt.show()
```

```python
x2=x+0.5
```

```python
plt.bar(x,sales_swift,width=0.3,color='red')
plt.bar(x2,sales_nano,width=0.3,alpha=0.7)
plt.show()
```

```python
x3=x+0.3
```

```python
plt.bar(x,sales_swift,width=0.3,color='red')
plt.bar(x3,sales_nano,width=0.3,alpha=0.7)
plt.show()
```

```python
plt.bar(x,sales_swift,width=0.3,color='red',label='swift')
plt.bar(x3,sales_nano,width=0.3,alpha=0.7,label='nano')
plt.legend()
plt.xlabel("year")
plt.ylabel("Sales (Million Dollars)")
plt.show()
```

```python
x=np.arange(2017,2022,1)
x2=x+0.3
```

```python
plt.bar(x,sales_swift,width=0.3,color='red',label='swift')
plt.bar(x2,sales_nano,width=0.3,alpha=0.7,label='nano')
plt.legend()
plt.xlabel("year")
plt.ylabel("Sales (Million Dollars)")
plt.show()
```

```python
#Histogram
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marks=np.round(np.random.randn(100)*10+70) #generate 100 random numbers - multiply them with 10 - add 70 to them and take the round - so becomes a whole number
```

```python
marks
```

```python
plt.hist(marks)
#1st array is the count
#2nd array is the marks
```

```python
 plt.hist(marks)
#1st array is the count
#2nd array is the marks
plt.show()
```

```python
plt.hist(marks,bins=20)
#1st array is the count
#2nd array is the marks
plt.show()
```

```python
maths_marks=np.round(np.random.rand(500)*5+80)
```

```python
print(maths_marks[:10])
```

```python
physics_marks=np.round(np.random.randn(500)*5+70)
```

```python
print(physics_marks[:10])
```

```python
plt.hist(maths_marks,bins=20)
plt.hist(physics_marks,bins=20)
plt.show()
```

```python
plt.hist(maths_marks,bins=20)
plt.hist(physics_marks,bins=20,alpha=0.5)
plt.show()
```

```python
plt.hist(maths_marks,bins=20,label='Maths')
plt.hist(physics_marks,bins=20,alpha=0.5,label='Physics')
plt.legend()
plt.savefig('marks.png')
plt.show()
```
